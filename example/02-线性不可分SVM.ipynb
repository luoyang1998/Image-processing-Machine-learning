# 线性不可分支持向量机

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
```

```python
data = sio.loadmat('E:/ML_NG/6-SVM/data/ex6data2.mat')
data.keys()
```

```python
X,y = data['X'],data['y']
X.shape,y.shape
```

```python
def plot_data():
   plt.scatter(X[:,0],X[:,1],c = y.flatten(),cmap = 'jet')
   plt.xlabel('x1')
   plt.ylabel('y')
```

```python
plot_data()
```

```python
from sklearn.svm import SVC
```

```python
svc1 = SVC(C=1,kernel='rbf',gamma=50)
svc1.fit(X,y.flatten())
```

```python
svc1.score(X,y.flatten())
```

# 绘制决策边界

```python
def plot_boundary(model):
    x_min,x_max = 0,1
    y_min,y_max = 0.4,1
    xx,yy = np.meshgrid(np.linspace(x_min,x_max,500),
                      np.linspace(y_min,y_max,500))
    z = model.predict(np.c_[xx.flatten(),yy.flatten()])
                        
                        
    zz = z.reshape(xx.shape)
    plt.contour(xx,yy,zz)
```

```python
plot_boundary(svc1)
plot_data()
```
