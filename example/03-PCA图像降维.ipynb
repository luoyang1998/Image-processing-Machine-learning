# 案例：使用PCA进行图像降维，k=36

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
```

```python
mat = sio.loadmat('E:/ML_NG/7-kmeans and PCA/data/ex7faces.mat')
X = mat['X']
print(X.shape)
```

```python
def plot_100_images(X):

    fig,axs = plt.subplots(ncols=10,nrows=10,figsize=(10,10)) 
    
    for c in range(10):
        for r in range(10):
            axs[c,r].imshow(X[10 * c + r].reshape(32,32).T,cmap='Greys_r') #显示单通道灰度图
            axs[c,r].set_xticks([])
            axs[c,r].set_yticks([])
```

```python
plot_100_images(X)
```

```python
means = np.mean(X,axis=0)
X_demean = X - means
C= X_demean.T @ X_demean
U,S,V = np.linalg.svd(C)
```

```python
U1 = U[:,:36]
X_reduction = X_demean@U1
X_reduction.shape
```

```python
X_recover = X_reduction@U1.T + means
```

```python
plot_100_images(X_recover),plot_100_images(X)
```
